```python
sns.countplot(x="Class", data=df)
plt.title("Normal vs Fraud Transactions")
plt.xticks([0,1], ["Normal", "Fraud"])
plt.show()
```

```python
df["Amount"].describe()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["Amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()
### Transaction Amount Distribution

# The distribution of transaction amounts is highly skewed to the right.
# Most transactions involve small amounts, while very large transactions
# are rare. This pattern is typical in financial transaction datasets.
```
